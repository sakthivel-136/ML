```python
le=LabelEncoder()
sald['Education Level'] = le.fit_transform(sald['Education Level'])
sald['Gender'] = le.fit_transform(sald['Gender'])
sald.head()
```

```python
ind = sald[['Age', 'Gender', 'Education Level', 'Years of Experience']]
dep = sald['Salary']
```

```python
LR = LinearRegression()
LR.fit(ind, dep)
LinearRegression()
```

```python
Q=LR.predict([[21,0,2,5]])
print("SALARY IS",Q)
```
